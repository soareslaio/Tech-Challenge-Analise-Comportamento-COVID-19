```python
# Fun√ß√£o para verificar se pelo menos um valor √© igual a 1 em qualquer coluna
def verifica_rendimento(row):
    if 1 in row.values:
        return 1
    else:
        return 0

# Aplicando a fun√ß√£o para cada linha (axis=1) e criando a nova coluna 'recebe_rendimento'
df2['recebe_rendimento'] = df2.apply(verifica_rendimento, axis=1)
df1['recebe_rendimento'] = df1.apply(verifica_rendimento, axis=1)
df['recebe_rendimento'] = df.apply(verifica_rendimento, axis=1)
```

```python
import pandas as pd

# Definir as colunas de interesse
columns_of_interest = [
    'Somatorio_valores_aposentadoria_e_penssao',
    'Somatorio_bolsa_familia',
    'Somatorio_BPC_LOAS',
    'Somatorio_auxilio_emergencial_covid',
    'Somatorio_seguro_desemprego'
]

# Fun√ß√£o para calcular o rendimento m√©dio por pessoa
def calculate_average_income(df):
    # Selecionar e substituir valores nulos por 0 nas colunas de interesse
    df[columns_of_interest] = df[columns_of_interest].fillna(0)

    # Calcular o rendimento m√©dio por pessoa (soma dos valores de cada linha)
    df['rendimento_social_total_medio'] = df[columns_of_interest].sum(axis=1)

    return df

# Aplicar a fun√ß√£o aos DataFrames df1, df2 e df
df1 = calculate_average_income(df1)
df2 = calculate_average_income(df2)
df = calculate_average_income(df)
```

```python
# Definir as colunas de interesse
columns_of_interest = [
    'Somatorio_valores_aposentadoria_e_penssao',
    'Somatorio_bolsa_familia',
    'Somatorio_BPC_LOAS',
    'Somatorio_auxilio_emergencial_covid',
    'Somatorio_seguro_desemprego'
]

# Fun√ß√£o para calcular a nova coluna 'recebeu_n_rendimento_social'
def calculate_received_social_income(df):
    # Contar o n√∫mero de colunas com valor maior que zero para cada linha
    received_count = (df[columns_of_interest] > 0).sum(axis=1)

    # Aplicar condi√ß√µes para atribuir o valor √† nova coluna 'recebeu_n_rendimento_social'
    df['recebeu_n_rendimento_social'] = 0  # Inicializar com 0 por padr√£o

    for index, count in received_count.items():
        if count == 1:
            df.at[index, 'recebeu_n_rendimento_social'] = 1
        elif count > 1 and count <= 6:
            df.at[index, 'recebeu_n_rendimento_social'] = count
        # Caso count seja 0, manter o valor 0 que j√° foi atribu√≠do

    return df

# Aplicar a fun√ß√£o aos DataFrames df1, df2 e df (substitua df1, df2 e df pelos seus DataFrames)
df1 = calculate_received_social_income(df1)
df2 = calculate_received_social_income(df2)
df = calculate_received_social_income(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Columns of interest
columns_of_interest = ['recebe_rendimento']

# Function to count "1" occurrences in each column of interest
def count_ones(df):
    return df[columns_of_interest].eq(1).sum()

# Count "1" occurrences in the columns of interest for each DataFrame
count_df2 = count_ones(df2)
count_df1 = count_ones(df1)
count_df = count_ones(df)

# Create DataFrame with the counts
df_counts = pd.DataFrame({
    'M√™s': ['Setembro/20', 'Outubro/20', 'Novembro/20'],
    'N√∫mero de Respostas': [count_df2['recebe_rendimento'], count_df1['recebe_rendimento'], count_df['recebe_rendimento']],
})

# Set seaborn style and color palette
sns.set(style='whitegrid')

# Plot line chart for monthly evolution of positive symptom cases
plt.figure(figsize=(10, 6))

# Loop over symptom columns (except 'Month') to plot each line in the chart
for column in df_counts.columns[1:]:  # Start from the second element to avoid 'Month'
    sns.lineplot(data=df_counts, x='M√™s', y=column, label=None, marker='o', palette='Rd')

    # Add text with the quantity and percentage above each point
    for i, value in enumerate(df_counts[column]):
        # Calculate percentage representation
        percentage = round((value / df.shape[0]) * 100, 1)  # Get total rows from df

        # Add annotation with value and percentage
        plt.annotate(f"({percentage}%)", (i, value), textcoords="offset points", xytext=(10, 5), ha='center', fontsize=8, color='black')

# Add title, labels, and legend
plt.title('Recebeu pelo Menos Um Tipo de Benef√≠cio Social (Percentual do Total de Entrevistados)', fontsize=14)
plt.xlabel('M√™s', fontsize=10)
plt.ylabel('N√∫mero de Respostas', fontsize=10)
plt.ylim(0, df_counts['N√∫mero de Respostas'].max() * 1.1)  # Set y-axis limit starting from 0
plt.grid(False)

# Display the chart
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Fun√ß√£o para adicionar r√≥tulos com porcentagens acima das barras
def add_percentage_labels(ax, labels, counts):
    total = sum(counts)
    for i, label in enumerate(labels):
        percentage = (counts[i] / total) * 100
        ax.text(i, counts[i] + 5, f'{percentage:.2f}%', ha='center')

# Calcular value_counts() para cada DataFrame
vc_set20 = df2['recebeu_n_rendimento_social'].value_counts()
vc_out20 = df1['recebeu_n_rendimento_social'].value_counts()
vc_nov20 = df['recebeu_n_rendimento_social'].value_counts()

# R√≥tulos para cada categoria (valores √∫nicos ordenados)
labels = sorted(df2['recebeu_n_rendimento_social'].unique())

# Plotar os gr√°ficos de barra
plt.figure(figsize=(15, 5))

# Gr√°fico para Set/20
ax1 = plt.subplot(1, 3, 1)
bars1 = ax1.bar(labels, vc_set20[labels], color='skyblue')
plt.title('Set/20 - Recebimento de Benef√≠cio Social')
plt.xlabel('Qtd. de Rendimentos Distintos')
plt.ylabel('Quantidade de Entrevistado')
ax1.grid(False)  # Desativar linhas de grade
add_percentage_labels(ax1, labels, vc_set20[labels])

# Gr√°fico para Out/20
ax2 = plt.subplot(1, 3, 2)
bars2 = ax2.bar(labels, vc_out20[labels], color='salmon')
plt.title('Out/20 - Recebimento de Benef√≠cio Social')
plt.xlabel('Qtd. de Rendimentos Distintos')
plt.ylabel('Quantidade de Entrevistado')
ax2.grid(False)  # Desativar linhas de grade
add_percentage_labels(ax2, labels, vc_out20[labels])

# Gr√°fico para Nov/20
ax3 = plt.subplot(1, 3, 3)
bars3 = ax3.bar(labels, vc_nov20[labels], color='lightgreen')
plt.title('Nov/20 - Recebimento de Benef√≠cio Social')
plt.xlabel('Qtd. de Rendimentos Distintos')
plt.ylabel('Quantidade de Entrevistado')
ax3.grid(False)  # Desativar linhas de grade
add_percentage_labels(ax3, labels, vc_nov20[labels])

# Ajustando o layout e exibindo os gr√°ficos
plt.tight_layout()
plt.show()
```

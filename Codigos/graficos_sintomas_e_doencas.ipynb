```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Definir os DataFrames df2 (Setembro), df1 (Outubro) e df (Novembro)
# Substitua df2, df1 e df pelos seus DataFrames reais

# Selecionar a coluna de interesse (medida de restrição de contato com pessoas)
column_of_interest = 'medida_restricao_contato_pessoas'

# Mapeamento de valores para descrições correspondentes
mapping = {
    1: 'Não fez restrição, levou vida normal como antes da pandemia',
    2: 'Reduziu o contato com as pessoas, mas continuou saindo de casa para trabalho ou atividades não essenciais e/ou recebendo visitas',
    3: 'Ficou em casa e só saiu em caso de necessidade básica',
    4: 'Ficou rigorosamente em casa',
    9: 'Ignorado'
}

# Função para contar ocorrências de cada valor mapeado na coluna de interesse
def count_restrictions(df):
    # Substituir valores na coluna de interesse pelo mapeamento
    restricted_column = df[column_of_interest].map(mapping)

    # Contar ocorrências de cada descrição
    value_counts = restricted_column.value_counts().reindex(mapping.values(), fill_value=0)

    return value_counts

# Contar ocorrências das medidas de restrição para cada DataFrame
count_df2 = count_restrictions(df2)
count_df1 = count_restrictions(df1)
count_df = count_restrictions(df)

# Criar DataFrame com as contagens
df_counts = pd.DataFrame({
    'Mês': ['Setembro/20', 'Outubro/20', 'Novembro/20'],
    'Não fez Restrição': [count_df2['Não fez restrição, levou vida normal como antes da pandemia'], count_df1['Não fez restrição, levou vida normal como antes da pandemia'], count_df['Não fez restrição, levou vida normal como antes da pandemia']],
    'Reduziu o Contato, mas Saiu para Trabalho ou Atividade Não Essenciais': [count_df2['Reduziu o contato com as pessoas, mas continuou saindo de casa para trabalho ou atividades não essenciais e/ou recebendo visitas'], count_df1['Reduziu o contato com as pessoas, mas continuou saindo de casa para trabalho ou atividades não essenciais e/ou recebendo visitas'], count_df['Reduziu o contato com as pessoas, mas continuou saindo de casa para trabalho ou atividades não essenciais e/ou recebendo visitas']],
    'Ficou em casa e Saiu só para Necessidades Básicas': [count_df2['Ficou em casa e só saiu em caso de necessidade básica'], count_df1['Ficou em casa e só saiu em caso de necessidade básica'], count_df['Ficou em casa e só saiu em caso de necessidade básica']],
    'Ficou Rigorosamente em Casa': [count_df2['Ficou rigorosamente em casa'], count_df1['Ficou rigorosamente em casa'], count_df['Ficou rigorosamente em casa']],
    'Ignorou a Pergunta': [count_df2['Ignorado'], count_df1['Ignorado'], count_df['Ignorado']]
})

# Configurar estilo seaborn e paleta de cores
sns.set(style='whitegrid')

# Plotar gráfico de linhas para evolução mensal das medidas de restrição
plt.figure(figsize=(12, 8))

# Loop sobre as colunas de medidas de restrição para plotar cada linha no gráfico
for column in df_counts.columns[1:]:
    sns.lineplot(data=df_counts, x='Mês', y=column, label=column, marker='o')

    # Adicionar rótulos com quantidade e percentual acima das bolinhas
    for i, value in enumerate(df_counts[column]):
        total_responses_month = df_counts.iloc[i, 1:].sum()  # Soma total das respostas para o mês atual
        percent_value = (value / total_responses_month) * 100

        # Obtém o valor formatado para exibir acima da bolinha
        value_formatted = f'{value} ({percent_value:.0f}%)'

        # Adiciona o texto acima da bolinha
        plt.text(i, value + 1200, value_formatted, ha='center', fontsize=9)

# Adicionar título, rótulos e legenda
plt.title('Na semana passada, devido à pandemia do Coronavírus, em que medida restringiu o contato com as pessoas?', fontsize=14)
plt.xlabel('Mês', fontsize=12)
plt.ylabel('Número de Respostas', fontsize=12)
plt.legend(title='Medida de Restrição', fontsize=10)
plt.grid(False)

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

```python
def colunas_diferentes_dfs(dfs):
    # Verificar se há pelo menos dois DataFrames
    if len(dfs) < 2:
        return "Não há DataFrames suficientes para comparar."

    # Obter conjunto de colunas do primeiro DataFrame
    colunas_base = set(dfs[0].columns)
    colunas_diferentes = []

    # Verificar as colunas diferentes em relação ao primeiro DataFrame
    for i, df in enumerate(dfs[1:], start=1):
        colunas_dif = set(df.columns) - colunas_base
        if colunas_dif:
            for coluna in colunas_dif:
                colunas_diferentes.append((coluna, f"df{i+1}"))

    if colunas_diferentes:
        return "Não", colunas_diferentes
    else:
        return "Sim", None

# Exemplo de uso
resultado, colunas_diferentes = colunas_diferentes_dfs([df, df1, df2])

if resultado == "Sim":
    print("As colunas dos DataFrames são iguais.")
else:
    print("As colunas dos DataFrames são diferentes. Colunas diferentes:")
    for coluna, df_nome in colunas_diferentes:
        print(f"Coluna '{coluna}' existe no {df_nome}.")
```

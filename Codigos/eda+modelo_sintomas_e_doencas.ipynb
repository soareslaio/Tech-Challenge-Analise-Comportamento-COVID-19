```python
def colunas_diferentes_dfs(dfs):
    # Verificar se há pelo menos dois DataFrames
    if len(dfs) < 2:
        return "Não há DataFrames suficientes para comparar."

    # Obter conjunto de colunas do primeiro DataFrame
    colunas_base = set(dfs[0].columns)
    colunas_diferentes = []

    # Verificar as colunas diferentes em relação ao primeiro DataFrame
    for i, df in enumerate(dfs[1:], start=1):
        colunas_dif = set(df.columns) - colunas_base
        if colunas_dif:
            for coluna in colunas_dif:
                colunas_diferentes.append((coluna, f"df{i+1}"))

    if colunas_diferentes:
        return "Não", colunas_diferentes
    else:
        return "Sim", None

# Exemplo de uso
resultado, colunas_diferentes = colunas_diferentes_dfs([df, df1, df2])

if resultado == "Sim":
    print("As colunas dos DataFrames são iguais.")
else:
    print("As colunas dos DataFrames são diferentes. Colunas diferentes:")
    for coluna, df_nome in colunas_diferentes:
        print(f"Coluna '{coluna}' existe no {df_nome}.")
```

```python
#O LabelEncoder é uma classe do scikit-learn utilizada para codificar rótulos de variáveis
#\n categóricas em valores numéricos. Em problemas de aprendizado de máquina, muitos algoritmos requerem que todas as entradas sejam numéricas,
#\n nportanto, é necessário converter variáveis categóricas (por exemplo, strings ou categorias) em números antes de alimentar os dados em um modelo.

#A função principal do LabelEncoder é mapear cada categoria única em uma coluna para um número
#\n inteiro único. Por exemplo, se uma coluna contém categorias como "maçã", "laranja" e "banana",
#\n o LabelEncoder atribuirá um número único a cada uma delas, como 0 para "maçã", 1 para "laranja" e 2 para "banana".
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score

# Dividindo os dados em conjunto de treinamento e teste
x_train, x_test, y_train, y_test = train_test_split(x_equilibrado, y_equilibrado, test_size=0.3, stratify=y_equilibrado, random_state=7)

# Definindo uma grade de hiperparâmetros mais limitada
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# Instanciando o modelo RandomForestClassifier
rf = RandomForestClassifier(random_state=42)

# Configurando a busca em grade com validação cruzada
grid_search = GridSearchCV(rf, param_grid, cv=3, scoring='accuracy')

# Treinamento do modelo com a busca em grade
grid_search.fit(x_train, y_train)

# Melhor modelo encontrado pela busca em grade
best_rf = grid_search.best_estimator_

# Fazendo previsões no conjunto de teste com o melhor modelo
y_pred = best_rf.predict(x_test)
y_prob = best_rf.predict_proba(x_test)[:, 1]

# Calculando e imprimindo as métricas
accuracy = accuracy_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_prob)

print("Melhor Modelo RandomForestClassifier:")
print("Acurácia: {:.4f}".format(accuracy))
print("AUC: {:.4f}".format(auc))
print("Melhores Parâmetros:", grid_search.best_params_)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, roc_auc_score

# Dividindo os dados em conjunto de treinamento e teste
x_train, x_test, y_train, y_test = train_test_split(x_equilibrado, y_equilibrado, test_size=0.3, stratify=y_equilibrado, random_state=7)

# Escalonando os dados
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Definindo uma grade de hiperparâmetros para a Regressão Logística
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10],  # Parâmetro de regularização
    'penalty': ['l1', 'l2']  # Tipo de regularização (L1 ou L2)
}

# Instanciando o modelo LogisticRegression com max_iter aumentado
logreg = LogisticRegression(max_iter=1000, random_state=42)

# Configurando a busca em grade com validação cruzada
grid_search = GridSearchCV(logreg, param_grid, cv=3, scoring='accuracy')

# Treinamento do modelo com a busca em grade nos dados escalados
grid_search.fit(x_train_scaled, y_train)

# Melhor modelo encontrado pela busca em grade
best_logreg = grid_search.best_estimator_

# Fazendo previsões no conjunto de teste com o melhor modelo
y_pred = best_logreg.predict(x_test_scaled)
y_prob = best_logreg.predict_proba(x_test_scaled)[:, 1]

# Calculando e imprimindo as métricas
accuracy = accuracy_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_prob)

print("Melhor Modelo LogisticRegression:")
print("Acurácia: {:.4f}".format(accuracy))
print("AUC: {:.4f}".format(auc))
print("Melhores Parâmetros:", grid_search.best_params_)
```
